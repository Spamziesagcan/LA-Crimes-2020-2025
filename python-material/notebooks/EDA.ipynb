```python
df['Location'] = df['Location'].str.replace(r'\s+', ' ', regex=True)
df['AREA_NAME'] = df['AREA_NAME'].str.replace(r'\s+', ' ', regex=True)
df['Crm_Cd_Desc'] = df['Crm_Cd_Desc'].str.replace(r'\s+', ' ', regex=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(30,16))
sns.histplot(df['Crm_Cd_Desc'],kde='True',bins=30)
plt.xticks(rotation=90)
plt.title('Crime Code Description')
plt.xlabel('Crime Code Description')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(25,15))
sns.histplot(df['AREA_NAME'],kde='True',bins=30)
plt.xticks(rotation=90)
plt.title('Crime Locations')
plt.xlabel('Crime Area')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Vict_Age'],kde='True',bins=30)
plt.xticks(rotation=90)
plt.title('Victim Age Distribution')
plt.xlabel('Victim Age')
plt.ylabel('Frequency')
plt.show()
```

```python
crime_counts = df['Crm_Cd_Desc'].value_counts().head()

plt.figure(figsize=(6,6))
plt.pie(crime_counts, labels=crime_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Crime Types in Chicago')
plt.show()
```

```python
crime_counts = df['Vict_Sex'].value_counts().head()

plt.figure(figsize=(6,6))
plt.pie(crime_counts, labels=crime_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Victim Sex Distribution')
plt.show()
```

```python
crime_counts = df['Vict_Age'].value_counts().head(10)

plt.figure(figsize=(8,8))
plt.pie(crime_counts, labels=crime_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Victim Age Distribution')
plt.show()
```

```python
community_crime_counts = df.groupby('AREA_NAME').size().reset_index(name='crime_count')

plt.figure(figsize=(24,12))
plt.scatter(community_crime_counts['AREA_NAME'], community_crime_counts['crime_count'], color='teal')
plt.title('Crimes by Community Area')
plt.xlabel('Community Area')
plt.ylabel('Number of Crimes')
plt.grid(True)
plt.show()
```

```python
df['Date_Rptd'] = pd.to_datetime(df['Date_Rptd'], errors='coerce')

df['Year'] = df['Date_Rptd'].dt.year
df['Month'] = df['Date_Rptd'].dt.month
df['Day'] = df['Date_Rptd'].dt.day
```

```python
yearly_counts = df['Year'].value_counts().sort_index()

plt.figure(figsize=(10, 5))
plt.plot(yearly_counts.index, yearly_counts.values, marker='o')
plt.title('Crimes Reported Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Crimes')
plt.grid(True)
plt.show()
```
